```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

code for simple linear regression

```python
class LR_theory:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, X_train, y_train):
        num = 0
        den = 0
        for i in range(X_train.shape[0]):
            num = num + ((X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean()))
            den = den + ((X_train[i] - X_train.mean()) * ((X_train[i] - X_train.mean())))

        self.m = num / den
        self.b = y_train.mean() - (self.m * X_train.mean())
                    

        print(self.m)
        print(self.b)
    
    def predict(self, X_test):
        return self.m * X_test + self.b
    
```

```python
LR = LR_theory()
```

```python
LR.fit(X_train, y_train)
```

```python
l1 = []
for i in range (len(X_test)):
    l1.append(float(LR.predict(X_test[i])))
y_test_through_algo = np.array(l1)
```

```python
camparison =  pd.DataFrame({
    "answer through algo": y_test_through_algo,
    "y test OG values": y_test,
     "difference": y_test -  y_test_through_algo
})
```
